```python
q1=USAHousing.quantile(0.25)
q3=USAHousing.quantile(0.75)
IQR=q3-q1
pos_outlier=q3 + 1.5*IQR
neg_outlier=q1 - 1.5*IQR
```

```python
# Select only numeric columns for IQR calculation
numeric_cols = USAHousing.select_dtypes(include=[np.number]).columns
q1 = USAHousing[numeric_cols].quantile(0.25)
q3 = USAHousing[numeric_cols].quantile(0.75)
IQR = q3 - q1
pos_outlier = q3 + 1.5 * IQR
neg_outlier = q1 - 1.5 * IQR
```

```python
print(q1)
print(q3)
print(IQR)
print(pos_outlier)
print(neg_outlier)
```

```python
USAHousing = USAHousing[~((USAHousing[numeric_cols] < (q1 - 1.5 * IQR)) | 
                          (USAHousing[numeric_cols] > (q3 + 1.5 * IQR))).any(axis=1)]
```

```python
#Address is a non significant variable so we are dropping it
USAHousing=USAHousing.iloc[:,0:-1]
```
