```python
# This already gives you the encoded DataFrame - no need for OneHotEncoder
sep_col = pd.get_dummies(df[['mainroad','guestroom', 'basement', 'hotwaterheating', 
                           'airconditioning', 'prefarea', 'furnishingstatus']], drop_first=True)

print("Encoded columns:")
print(sep_col.columns.tolist())
print("\nFirst few rows:")
print(sep_col.head())

# sep_col is already your encoded DataFrame - use it directly
lt_cl = sep_col.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
encoded_data = ohe.fit_transform(df[['mainroad', 'guestroom', 'basement', 'hotwaterheating',
                                   'airconditioning', 'prefarea', 'furnishingstatus']])

# Simply convert to integer
lt_cl = pd.DataFrame(encoded_data.astype(int), columns=ohe.get_feature_names_out())

print(lt_cl.head())
print("\nUnique values:", lt_cl.values.flatten())
```

```python
lt_cl
```

```python
df_n=pd.concat([df,lt_cl],axis='columns')
```

```python
df_n
```

```python
df_n=df_n.drop(['mainroad', 'guestroom', 'basement', 'hotwaterheating','airconditioning', 'prefarea', 'furnishingstatus'],axis='columns')
df_n
```

```python
inp_var=df_n.drop(['price'],axis='columns')
target=df_n.price
```

```python
inp_var
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inp_var,target,test_size=0.3,random_state=70)
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
y_predict=linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
